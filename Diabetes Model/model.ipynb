```python
import tensorflow as tf 
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras import Sequential
import numpy as np
import pandas

print(tf.config.list_physical_devices())
```

```python
# EXTRACTING THE DATA 
diabetes = pandas.read_csv("diabetes.csv")

pregnancies = diabetes['Pregnancies'].to_numpy()
glucose = diabetes['Glucose'].to_numpy()
bp = diabetes['BloodPressure'].to_numpy()
skin_thickness = diabetes['SkinThickness'].to_numpy()
insulin = diabetes['Insulin'].to_numpy()
bmi = diabetes['BMI'].to_numpy()
dpf = diabetes['DiabetesPedigreeFunction'].to_numpy()
age = diabetes['Age'].to_numpy()

y_train = diabetes['Outcome'].to_numpy()
x_train = np.transpose(np.array([pregnancies, glucose, bp, skin_thickness, insulin, bmi, dpf, age]))
print(f"x training data: {x_train[0]}\n")
```

```python
# CREATE TESTING DATA 
x_test = np.array([ [6,148,72,35,0,33.6,0.627,50],
                    [1,85,66,29,0,26.6,0.351,31],
                    [8,183,64,0,0,23.3,0.672,32],
                    [1,89,66,23,94,28.1,0.167,21],
                    [0,137,40,35,168,43.1,2.288,33]])
y_test = np.array([1, 0, 1, 0, 1])
```

```python
model = Sequential([
    Dense(units=5, activation="sigmoid"),
    Dense(units=1, activation="sigmoid")
])

with tf.device('cpu'):
    model.compile(loss=BinaryCrossentropy())
    model.fit(x_train, y_train, epochs=3000)

```

```python
for i in range(len(x_test - 1)):
    print(f"Prediction: {round(model.predict(np.array([x_test[i]]))[0][0])}")
    print(f"Actual Result: {y_test[i]}")
```
